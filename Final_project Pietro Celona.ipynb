```python
import pandas as pd
```

```python
url='https://covid.ourworldindata.org/data/owid-covid-data.csv'
```

```python
df_covid_raw = pd.read_csv(url)
df_covid_raw
```

```python
continent_list = df_covid_raw['continent'].drop_duplicates().dropna()
```

```python
df_total_cases_per_continent = df_covid_raw[['continent','location' ,'total_cases']].dropna().groupby(['continent', 'location']).max().groupby('continent').sum()
df_total_cases_per_continent
```

```python
df_total_vaccinations_per_continent = df_covid_raw[['continent','location' ,'total_vaccinations']].dropna().groupby(['continent', 'location']).max().groupby('continent').sum()
df_total_vaccinations_per_continent
```

```python
def compare_continents_covid_cases(df: pd.DataFrame, continents:list):
    
    total_cases = df_total_cases_per_continent.sum()

    continent_confront = df['continent'].isin(continents)

    df_filtered_per_location = df[continent_confront].loc[:,['continent', 'location','total_cases']].dropna().groupby(['continent', 'location'])['total_cases']
    df_filtered_per_continent = df[continent_confront].loc[:,['continent', 'location','total_cases']].dropna().groupby(['continent'])['total_cases']

    df_with_useful_values_per_location = df_filtered_per_location.agg(['min', 'max']).reset_index()
    df_with_useful_values_per_continent = df_filtered_per_continent.agg(['min', 'max'])
    
    df_with_useful_values_per_location['precentVsTotal'] = df_with_useful_values_per_location['max'].apply(lambda x: (x/total_cases))
    
    df_with_useful_values_per_continent['mean_max'] = df_with_useful_values_per_location.groupby(['continent'])['max'].mean()
    df_with_useful_values_per_continent['mean_min'] = df_with_useful_values_per_location.groupby(['continent'])['min'].mean()
    
    df_with_useful_values_per_continent['precentVsTotal'] = df_with_useful_values_per_location.groupby(['continent'])['max'].sum().apply(lambda x: (x/total_cases) * 100)

    print(df_with_useful_values_per_continent.reset_index())
    return(df_with_useful_values_per_location)
```

```python
def compare_continents_vaccinations(df: pd.DataFrame, continents:list):
    
    total_vaccinations = df_covid_raw[['continent','location' ,'total_vaccinations']].dropna().groupby(['continent', 'location']).max().sum()

    continent_confront = df['continent'].isin(continents)

    df_filtered_per_location = df[continent_confront].loc[:,['continent', 'location','total_vaccinations']].dropna().groupby(['continent', 'location'])['total_vaccinations']
    df_filtered_per_continent = df[continent_confront].loc[:,['continent', 'location','total_vaccinations']].dropna().groupby(['continent'])['total_vaccinations']

    df_with_useful_values_per_location = df_filtered_per_location.agg(['min', 'max']).reset_index()
    df_with_useful_values_per_continent = df_filtered_per_continent.agg(['min', 'max'])
    
    df_with_useful_values_per_location['precentVsTotal'] = df_with_useful_values_per_location['max'].apply(lambda x: (x/total_vaccinations))
    
    df_with_useful_values_per_continent['mean_max'] = df_with_useful_values_per_location.groupby(['continent'])['max'].mean()
    df_with_useful_values_per_continent['mean_min'] = df_with_useful_values_per_location.groupby(['continent'])['min'].mean()

    df_with_useful_values_per_continent['precentVsTotal'] = df_with_useful_values_per_location.groupby(['continent'])['max'].sum().apply(lambda x: (x/total_vaccinations) * 100)

    print(df_with_useful_values_per_continent.reset_index())

    return(df_with_useful_values_per_location)
```

```python
compare_continents_covid_cases(df_covid_raw, ['Europe', 'South America', 'Oceania'])
```

```python
compare_continents_vaccinations(df_covid_raw, ['Europe', 'South America', 'Oceania'])
```

L'europa ha il 32% dei casi totali registrati e l'11% dei delle vaccinazioni totali. Quindi ha il numero totale di casi e vaccinazioni più alti rispetto ai 3 continenti presi in analisi.
Il continente con meno casi e vaccinazioni è l'oceania che vanta circa l'2% dei casi totali di covid e meno dell'1% delle vaccinazioni totali
Il sud-america ha il circa il 9% dei casi totali e il 7% delle vaccinazioni totali, una nota riguardo il sud America è che la media dei massimi dei casi totali per locazione è maggiore rispetto a quella dell'europa, mentre la media massima di vaccinazioni per locazione, è molto più alta di quella europea.
